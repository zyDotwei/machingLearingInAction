## scikit-learn 中的逻辑回归

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(666)
X = np.random.normal(0, 1, size=(200, 2))
y = np.array(X[:,0] ** 2 + X[:,1] < 1.5, dtype='int')
```

```python
for i in range(20):
    y[np.random.randint(200)] = 1
```

```python
plt.scatter(X[y==0, 0], X[y==0, 1])
plt.scatter(X[y==1, 0], X[y==1, 1])
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=666)
```

### 使用sklearn

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
log_reg.score(X_train, y_train)
```

```python
log_reg.score(X_test, y_test)
```

```python
def plot_decision_boundary(model,axis):#绘图模块      
    x0,x1 = np.meshgrid(            
        np.linspace(axis[0], axis[1], int((axis[1] - axis[0]) * 100)).reshape(-1, 1),  
        np.linspace(axis[2], axis[3], int((axis[3] - axis[2]) * 100)).reshape(-1, 1) 
    )      
    # np.c_ 按列叠加两个矩阵的意思，也可以说是按行连接两个矩阵，就是把两矩阵左右相加，要求行数相等
    x_new = np.c_[x0.ravel(),x1.ravel()]   # ravel() 降维操作
    
    y_predict = model.predict(x_new)   
    zz = y_predict.reshape(x0.shape)    
    
    from matplotlib.colors import ListedColormap    
    custom_cmap = ListedColormap(['#EF9A9A', '#FFF59D','#90CAF9'])    
    plt.contourf(x0, x1, zz, linewidth=5, cmap=custom_cmap) 
```

```python
plot_decision_boundary(log_reg, axis=[-4, 4, -4, 4])
plt.scatter(X[y==0, 0], X[y==0, 1])
plt.scatter(X[y==1, 0], X[y==1, 1])
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler

def PolynormialLogisticRegression(degrees):
    return Pipeline([
        ('poly', PolynomialFeatures(degree=degrees)),
        ('std_scalar', StandardScaler()),
        ('log_reg', LogisticRegression())
    ])
```

```python
poly_log_reg = PolynormialLogisticRegression(degrees=2)
poly_log_reg.fit(X_train, y_train)
```

```python
poly_log_reg.score(X_train, y_train)
```

```python
poly_log_reg.score(X_test, y_test)
```

```python
plot_decision_boundary(poly_log_reg, axis=[-4, 4, -6, 4])
plt.scatter(X[y==0, 0], X[y==0, 1])
plt.scatter(X[y==1, 0], X[y==1, 1])
plt.show()
```

```python
poly_log_reg2 = PolynormialLogisticRegression(degrees=20)
poly_log_reg2.fit(X_train, y_train)
```

```python
poly_log_reg2.score(X_train, y_train)
```

```python
poly_log_reg2.score(X_test, y_test)
```

```python
plot_decision_boundary(poly_log_reg2, axis=[-4, 4, -6, 4])
plt.scatter(X[y==0, 0], X[y==0, 1])
plt.scatter(X[y==1, 0], X[y==1, 1])
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler

def PolynormialLogisticRegression(degrees=2,C=1):
    return Pipeline([
        ('poly', PolynomialFeatures(degree=degrees)),
        ('std_scalar', StandardScaler()),
        ('log_reg', LogisticRegression(C=C))
    ])
```

```python
poly_log_reg3 = PolynormialLogisticRegression(degrees=20, C=0.1)
poly_log_reg3.fit(X_train, y_train)
```

```python
poly_log_reg3.score(X_train, y_train)
```

```python
poly_log_reg3.score(X_test, y_test)
```

```python
plot_decision_boundary(poly_log_reg3, axis=[-4, 4, -6, 4])
plt.scatter(X[y==0, 0], X[y==0, 1])
plt.scatter(X[y==1, 0], X[y==1, 1])
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler

def PolynormialLogisticRegression(degrees=2, C=1, penalty='l2'):
    return Pipeline([
        ('poly', PolynomialFeatures(degree=degrees)),
        ('std_scalar', StandardScaler()),
        ('log_reg', LogisticRegression(C=C, penalty=penalty))
    ])
```

```python
poly_log_reg4 = PolynormialLogisticRegression(degrees=20, C=0.1, penalty='l1')
poly_log_reg4.fit(X_train, y_train)
```

```python
poly_log_reg4.score(X_train, y_train)
```

```python
poly_log_reg4.score(X_test, y_test)
```

```python
plot_decision_boundary(poly_log_reg4, axis=[-4, 4, -6, 4])
plt.scatter(X[y==0, 0], X[y==0, 1])
plt.scatter(X[y==1, 0], X[y==1, 1])
plt.show()
```
