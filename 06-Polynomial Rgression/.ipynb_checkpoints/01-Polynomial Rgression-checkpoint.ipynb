## 多项式回归

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(-3, 3, size=100)  # 均匀分布
X = x.reshape(-1, 1)
```

```python
y = 0.5 * x ** 2 + x + 2 + np.random.normal(0, 1, size=100)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
y_predict = lin_reg.predict(X)
```

```python
plt.scatter(x,y)
plt.plot(x, y_predict, color='r')
plt.show()
```

## #解决方案：添加一个特征

```python
(X**2).shape
```

```python
X2 = np.hstack([X, X**2])
```

```python
X2.shape
```

```python
X2[:10, :]
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(X2, y)
```

```python
y_predict2 = lin_reg2.predict(X2)
```

```python
plt.scatter(x,y)
plt.plot(x, y_predict2, color='r')
plt.show()
```

```python
plt.scatter(x,y)
plt.plot(np.sort(x), y_predict2[np.argsort(x)], color='r')
plt.show()
```

```python
lin_reg2.score(X2, y)
```
