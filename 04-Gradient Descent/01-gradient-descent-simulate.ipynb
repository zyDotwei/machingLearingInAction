```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plot_x = np.linspace(-1, 6, 141)
plot_x
```

```python
plot_y = (plot_x - 2.5) ** 2 - 1 
```

```python
plt.plot(plot_x, plot_y)
```

```python
def DJ(theta):
    return 2*(theta - 2.5)
```

```python
def J(theta):
    return (theta - 2.5) ** 2 - 1 
```

```python
theta = 0.0
eta = 0.1
epsilon = 1e-8

while True:
    gradient = DJ(theta)
    last_theta = theta
    theta = theta - eta * gradient
    if (abs(J(theta) - J(last_theta)) < epsilon):
        break

print(theta)
print(J(theta))
```

```python
theta = 0.0
eta = 0.1
epsilon = 1e-8
theta_his = [theta]

while True:
    gradient = DJ(theta)
    last_theta = theta
    theta = theta - eta * gradient
    theta_his.append(theta)
    if (abs(J(theta) - J(last_theta)) < epsilon):
        break
```

```python
plt.plot(plot_x, plot_y)
plt.plot(np.array(theta_his),J(np.array(theta_his)),'r-o')
plt.show()
```

```python
def gradient_descent(int_theta, eta, epsilon=1e-8):
    theta = int_theta
    theta_his.clear()
    theta_his.append(int_theta)
    
    while True:
        gradient = DJ(theta)
        last_theta = theta
        theta = theta - eta * gradient
        theta_his.append(theta)
        if (abs(J(theta) - J(last_theta)) < epsilon):
            break

def plot_theta_his():
    plt.plot(plot_x, plot_y)
    plt.plot(np.array(theta_his),J(np.array(theta_his)),'r-o', markersize=3.5)
    plt.show()
```

```python
eta = 0.01
gradient_descent(0., eta)
plot_theta_his()
```

```python
def J(theta):
    try:
        return (theta - 2.5) ** 2 - 1 
    except:
        return float('inf')
```

```python
def gradient_descent(int_theta, eta, n_iters=1e4, epsilon=1e-8):
    theta = int_theta
    theta_his.clear()
    theta_his.append(int_theta)
    i_ters = 0
    
    while i_ters < n_iters:
        gradient = DJ(theta)
        last_theta = theta
        theta = theta - eta * gradient
        theta_his.append(theta)
        if (abs(J(theta) - J(last_theta)) < epsilon):
            break
        i_ters += 1
```

```python
eta = 1.1
gradient_descent(0., eta)
plot_theta_his()
```
